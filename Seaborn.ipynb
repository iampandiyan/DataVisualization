#swarmplot
 It creates a categorical scatterplot where data points are adjusted along the categorical axis to avoid overlapping.
 Non-overlapping points: 
 The points are jittered along the categorical axis to prevent overlapping, ensuring that the distribution of values is clear.
Categorical grouping: 
The plot groups data points based on categories, allowing comparisons across groups.
Combination with other plots: 
Swarmplots can be layered over boxplots, violin plots, or bar plots to provide detailed insights about the data distribution.


```python
sb.swarmplot(x='ssc_b',y='salary',data=dataset)
```

```python
sb.swarmplot(x='ssc_b',y='salary',data=dataset,hue='gender')
```
